This notebook contains LDA topic modeling using preselected high tf-idf tokens


```python
import pandas as pd
import gensim
import pickle
from gensim.models import CoherenceModel


n_topics = 10

with open('dictionary_r.pkl', 'rb') as fp:
    dictionary_r = pickle.load(fp)
# dictionary_r.filter_extremes(no_below=10, no_above=0.5)
    
with open('corpus_r.pkl', 'rb') as fp:
    corpus_r = pickle.load(fp)
```

```python
with open('../indus_wanted.pkl', 'rb') as fp:
    indus_wanted = pickle.load(fp)
```

```python
from gensim.models import TfidfModel
```

```python
model = TfidfModel(corpus_r)
```

```python
corpus_tfidf = [model[i] for i in corpus_r]
```

```python
corpus_wanted_tfidf = [corpus_tfidf[i] for i in indus_wanted.index]
```

```python
corpus_wanted_bow = [corpus_r[i] for i in indus_wanted.index]
```

```python
len(corpus_wanted_bow) == len(corpus_wanted_tfidf)
```

```python
len(corpus_wanted_bow)
```

```python
# chang to list of dicts
corpus_tfidfs = [] 
for corpus in corpus_wanted_tfidf:
    tfidf_dict = {}
    for tup in corpus:
        tfidf_dict[tup[0]] = tup[1]
    corpus_tfidfs.append(tfidf_dict)
        
```

```python
corpus_bows = [] 
for corpus in corpus_wanted_bow:
    bow_dict = {}
    for tup in corpus:
        bow_dict[tup[0]] = tup[1]
    corpus_bows.append(bow_dict)
       
```

```python
corpus_tfidfs[0].keys() == corpus_bows[0].keys()
```

```python
corpus_bows[0]
```

```python
def create_bow_by_tfidf(tfidf_dict, bow_dict):
    
    tfidf_mean = sum(tfidf_dict.values())/len(tfidf_dict)
    
    bow = []
    for k, v in tfidf_dict.items():
        if v > tfidf_mean:
            bow.append((k, bow_dict[k]))
    return bow
   
    
```

```python
final_bow = [ create_bow_by_tfidf(corpus_tfidfs[i], corpus_bows[i])   for i in range(16411)]
```

```python
with open('dictionary_r.pkl', 'rb') as fp:
    dictionary_r = pickle.load(fp)
```

```python
len(dictionary_r)
```

```python
%%time
senlda_r_4 = gensim.models.ldamodel.LdaModel(corpus=final_bow, id2word=dictionary_r, num_topics=4, alpha='auto', eta='auto')

#senlda_r.save('Models/Model_10')
```
